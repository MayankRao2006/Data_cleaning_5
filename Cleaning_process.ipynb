```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DF = pd.read_csv("messy_data.csv")
df = DF.copy()
df
```

```python
df.isnull().sum()
```

```python
df["Customer Name"]=df["Customer Name"].fillna(df["Email"])
```

```python
df.loc[df["Customer Name"].str.strip().str.lower() == 'missing', "Customer Name"] = "NaN"
```

```python
df
```

```python
df.replace({"NaN":"-", "missing":"-"})
df.fillna("-")
```

```python
df.loc[df["Email"].str.strip().str.lower() == "nan", "Email"] = "-"
df.loc[df["Email"].str.strip().str.lower() == "missing", "Email"] = "-"
```

```python
df
```

```python
df["Customer Name"] = df["Customer Name"].str.strip().str.split("@", regex=False).str[0]
```

```python
df.loc[df["Customer Name"].str.strip().str.lower() == "nan", "Customer Name"] = "-"
```

```python
df
```

```python
df["Customer Name"] = df["Customer Name"].apply(lambda x: "BOB" if x in ["BOB", "Bob"] else x.title())
```

```python
df
```

```python
df.loc[df["Email"].str.strip() == "-", "Email"] = df["Customer Name"]
```

```python
df
```

```python
df["Email"] = df["Email"].replace({"Alice":"alice@gmail.com", "BOB":"bob@outlook", "Charlie":"charlie@yahoo.com"})
```

```python
df
```

```python
corrections = {
    "Charlie": "charlie@yahoo.com",
    "Alice": "alice@gmail.com",
    "BOB": "bob@outlook"
}
df["Email"] = df["Customer Name"].map(corrections).fillna(df["Email"])
```

```python
df
```

```python
df["Quantity"] = df["Quantity"].fillna("-")
```

```python
df.loc[df["Quantity"].str.strip().str.lower() == "four", "Quantity"] = 4
df.loc[df["Quantity"].str.strip().str.lower() == "one", "Quantity"] = 1
df.loc[df["Quantity"].str.strip().str.lower() == "two", "Quantity"] = 2
```

```python
df
```

```python
df["Price"] = df["Price"].fillna("-")
```

```python
df
```

```python
df["Price"] = df["Price"].replace({"fifty":50})
```

```python
df
```

```python
df[["Customer Name", "Email"]] = df[["Customer Name", "Email"]].replace({"-":"Unknown"})
```

```python
df
```

```python
df = df.query("Quantity != '-' and Price != '-'")
```

```python
df = df[["Customer Name", "Email", "Quantity", "Price", "Date"]]
```

```python
df
```

```python
df = df.reset_index()
```

```python
df = df[["Customer Name", "Email", "Quantity", "Price", "Date"]]
```

```python
df
```

```python
df["Quantity"] = df["Quantity"].map(int)
df["Price"] = df["Price"].map(int)
```

```python
df["Revenue"] = df["Quantity"] * df["Price"]
```

```python
df = df[["Date", "Customer Name", "Email", "Quantity", "Price", "Revenue"]]
```

```python
# This is our final data
df.to_csv("Cleaned_Data.csv", index=False)
```
